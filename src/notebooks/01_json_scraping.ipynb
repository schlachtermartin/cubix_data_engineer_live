```python
import json, os, pandas as pd
```

```python
file_path = "../../data/json_handling/spotify_playlist.json"

with open(file_path, "r") as json_file:
    data = json.load(json_file)
```

```python
tracklist = []
for track in data["contents"]["items"]:
    tracklist.append({"name": track["name"], "playcount": track["playCount"]})

tracklist
```

```python
tracklist_df = pd.DataFrame(tracklist)

tracklist_df["playcount"] = tracklist_df["playcount"].astype("int")
```

```python
print(tracklist_df["name"][30])
```

```python
sum_playcount = tracklist_df["playcount"].sum()

print(sum_playcount)
```

```python
song_name = tracklist_df["name"][tracklist_df.loc[(tracklist_df == tracklist_df["playcount"].min()).any(axis=1)].index[0]]
print(f"The lowest playcount is {song_name}.")
```
